```python
# ============================
# 1️⃣ Imports & Config
# ============================

import pandas as pd
from pathlib import Path

# Display settings
pd.set_option("display.max_rows", 200)
pd.set_option("display.max_columns", 100)

print("Libraries loaded successfully.")
```

```python
# ============================
# 2️⃣ Define CSV Path (⚠️CHANGE PATH)
# ============================

processed_file = Path("../data/processed/YOUR_FOLDER/YOUR_FILE.csv")

print("File exists:", processed_file.exists())

if not processed_file.exists():
    raise FileNotFoundError("Update the path to your processed CSV.")
```

```python
# ============================
# 3️⃣ Load Dataset
# ============================

df = pd.read_csv(processed_file)

print("CSV loaded successfully.")
print(f"Shape: {df.shape}")
```

```python
# ============================
# 4️⃣ Basic Structure Overview
# ============================

print("Columns:")
print(df.columns.tolist())

print("\nData types:")
display(df.dtypes)

print("\nFirst 5 rows:")
display(df.head())
```

```python
# ============================
# 5️⃣ Missing Values Analysis
# ============================

print("Missing values per column:")
display(df.isna().sum())

print("\nMissing values percentage:")
display((df.isna().mean() * 100).round(2))
```

```python
# ============================
# 6️⃣ Unique Values in a Column (⚠️CHANGE COLUMN NAME)
# ============================

column_name = "YOUR_COLUMN"

if column_name in df.columns:
    unique_vals = df[column_name].unique()
    print(f"\nUnique values in '{column_name}' ({len(unique_vals)} total):")
    print(unique_vals)
else:
    print("Column not found. Update column_name.")
```

```python
# ============================
# 7️⃣ Value Counts - Categorical Distribution (⚠️CHANGE COLUMN NAME)
# ============================

column_name = "YOUR_COLUMN"

if column_name in df.columns:
    display(df[column_name].value_counts())
else:
    print("Column not found.")
```

```python
# ============================
# 8️⃣ Year Range (if dataset has year column)
# ============================

if "year" in df.columns:
    print("Year range:", df["year"].min(), "to", df["year"].max())
else:
    print("No 'year' column found.")
```

```python
# ============================
# 9️⃣ Descriptive Statistics (⚠️CHANGE NUMERIC COLUMN)
# ============================

numeric_column = "YOUR_COLUMN"

if numeric_column in df.columns:
    display(df[numeric_column].describe())
else:
    print("Numeric column not found.")
```

```python
# ============================
# 🔟 Check EU27 Coverage (if country column exists)
# ============================

EU27_COUNTRIES = [
    'AT','BE','BG','CY','CZ','DE','DK','EE','EL','ES',
    'FI','FR','HR','HU','IE','IT','LT','LU','LV','MT',
    'NL','PL','PT','RO','SI','SK','SE'
]

if "country" in df.columns:
    present = set(df["country"].unique())
    missing = set(EU27_COUNTRIES) - present
    
    print("Countries present:", len(present))
    print("Missing EU countries:", missing)
else:
    print("No country column found.")
```
